```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Amazon marketplace - Referral fees.csv')
```

```python
df.head()
```

```python
cats=set(df['Category'])
```

```python
data={}
```

```python
for cat in cats:
    cdf=df[df['Category']==cat]
    fees={}
    if len(cdf)==1:
        cfees=list(df['Referral Fee Percentage'])
        fees[1000000000]=float(float(cfees[0][:len(cfees[0])-1])/100)
        data[cat]=fees
    else:
        break_points=list(df['Price Range (INR)'])
        cfees=list(df['Referral Fee Percentage'])
        prices=[]
        for point in break_points:
            curr=""
            for c in point:
                if c>='0' and c<='9':
                    curr+=c
                elif curr!="":
                    prices.append(int(curr))
                    curr=""
            
        if curr!="":
            prices.append(int(curr))
        
        prices=set(prices)
        prices=list(prices)
        prices.sort()
        cfees.sort()
        
        for i in range(len(prices)-1):
            fees[prices[i]]=float(float(cfees[i][:len(cfees[i])-1])/100)
        
        fees[1000000000]=float(float(cfees[-1][:len(cfees[-1])-1])/100)
        
        data[cat]=fees        
```

```python
data
```

```python
import json
```

```python
with open('brackets.json', 'w') as fp:
    json.dump(data, fp)
```
